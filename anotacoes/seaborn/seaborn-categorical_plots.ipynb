# Plots Categóricos
- factorplot
- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot

```python
import seaborn as sns
import numpy as np
sns.set(style='darkgrid') # altera os estilo dos plot
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

## sns.barplot
x define a categoria e y dados a serem calculados após agruado em função de cada categoria em x

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

## sns.countplot
Retorna um gráfico com a contagem de quanto dados existem para cada categoria que há no conjunto informado como *x*

```python
sns.countplot(x='sex', data=tips)
```

## sns.boxplot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

O parâmetro `hue` subdivide as categorias segundo o valor passado:

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex')
```

Exemplo de paleta de cores `rainbow`:

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

O atributo `orient` define a orientação de exibição do gráfico:

```python
sns.boxplot(x='total_bill', data=tips, orient='h')
```

## sns.violinplot

Exibe a distribuição dos dados, em `y`, usando KDE baseado na categoria passada para `x`.

```python
sns.violinplot(x='day', y='tip', data=tips)
```

Assim com o já apresentado, `hue` subdivide o gráfico baseado sem seu parâmetro.

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True)
```

## sns.stripplot

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

## sns.swarmplot

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

## combinando dois plots

```python
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
sns.violinplot(x='day', y='total_bill', data=tips)
```

## sns.factorplot
A função `factorplot` é capaz de criar outros plot através do que é definido no parâmetro `kind`

```python
sns.factorplot(x='sex', y='total_bill', data=tips, kind='bar')
```
